```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
df['doornumber']=le.fit_transform(df['doornumber'])
```

```python
df['carbody']=le.fit_transform(df['carbody'])
```

```python
df['aspiration']=le.fit_transform(df['aspiration'])
```

```python
X=df[['wheelbase', 'carlength', 'carwidth', 'carheight', 'curbweight',
       'enginesize', 'boreratio', 'stroke', 'compressionratio', 'horsepower',
       'peakrpm', 'citympg', 'highwaympg','aspiration', 'doornumber', 'carbody']].values
Y=df[['price']].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,train_size=0.8)
```

```python
X_train.shape, X_test.shape, Y_train.shape, Y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
lr.fit(X_train,Y_train)
```

```python
df[['wheelbase', 'carlength', 'carwidth', 'carheight', 'curbweight',
       'enginesize', 'boreratio', 'stroke', 'compressionratio', 'horsepower',
       'peakrpm', 'citympg', 'highwaympg','aspiration', 'doornumber', 'carbody','price']]
```

```python
lr.predict([[89,169,65,49,2550,131,3.45,2.78,9.1,112,5000,22,27,0,1,1]])
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error
```

```python
Y_pred=lr.predict(X_test)
```

```python
mean_squared_error(Y_pred,Y_test)
```

```python
mean_absolute_error(Y_pred,Y_test)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(Y_test,Y_pred)
```
